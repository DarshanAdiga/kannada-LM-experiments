```python
# iNLTK sentence_split
# https://indic-nlp-library.readthedocs.io/en/latest/indicnlp.tokenize.html
from indicnlp.tokenize import sentence_tokenize
def indicnlp_splitter(text):
    return sentence_tokenize.sentence_split(text, lang='kn')

indicnlp_splitter(text)
```

```python
# NLTK sent_tokenize
# http://www.nltk.org/api/nltk.tokenize.html#nltk.tokenize.sent_tokenize
from nltk.tokenize import sent_tokenize
def nltk_splitter(text):
    return sent_tokenize(text)

nltk_splitter(text)
```

```python
# Spacy Sentence Segmentation
# https://spacy.io/usage/linguistic-features#sbd
import spacy
from spacy.pipeline import Sentencizer
nlp = spacy.load("xx_ent_wiki_sm")
nlp.add_pipe('sentencizer')
# nlp

def spacy_splitter(text):
    doc = nlp(text)
    sents = []
    for sent in doc.sents:
        sents.append(sent.text)
    return sents

spacy_splitter(text)
```

```python
# Generate sentences
# NOTE: Ignoring the URLs here. If you want to keep track, do so here.
indic_s=[]
nltk_s=[]
spacy_s=[]
for text in sample_articles:
    indic_list = indicnlp_splitter(text)
    nltk_list = nltk_splitter(text)
    spacy_list = spacy_splitter(text)
    mx_len = max(len(indic_list), len(nltk_list), len(spacy_list))
    # Pad the list of sentences for this text
    if mx_len > len(indic_list):
        pad_len = mx_len - len(indic_list)
        indic_list = indic_list + ([None] * pad_len)
    if mx_len > len(nltk_list):
        pad_len = mx_len - len(nltk_list)
        nltk_list = nltk_list + ([None] * pad_len)
    if mx_len > len(spacy_list):
        pad_len = mx_len - len(spacy_list)
        spacy_list = spacy_list + ([None] * pad_len)

    indic_s = indic_s + indic_list
    nltk_s = nltk_s + nltk_list
    spacy_s = spacy_s + spacy_list
    
sent_columns = ['indic', 'nltk', 'spacy']
all_sents = {'indic': indic_s, 'nltk': nltk_s, 'spacy': spacy_s}
df_all_sents = pd.DataFrame(all_sents)
```
